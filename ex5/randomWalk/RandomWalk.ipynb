# Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Classes:

```python
class RandomWalk:
    """
    the class to simulate random Walk
    the walker randomly goes right with possibility p and left with possibility 1-p
    walks is the array containing random numbers representing walker's path

    methods:

    getX:
        the function to determine the position of the walker by sum of the steps the walker have done
    """
    def __init__(self, p: float, steps: int) -> None:
        """
        the function to construct RandomWalk object

        :param p: the possibility to go to the right(a number between 0 and 1)
        :type p: float
        :param steps: the total steps the walker goes
        :type steps: int
        """
        self.steps = steps
        self.p = p
        self.walks = 2*(np.ones(self.steps)*np.random.uniform(0,1,self.steps)<self.p) - np.ones(self.steps)

    def getX(self, t: int) -> np.ndarray:
        """
        the function to determine the position of the walker by sum of the steps the walker have done

        :param t: the input time we need to determine the position at
        :type t: int
        :return: the position of the walker
        :rtype: int
        """
        return np.sum(self.walks[:t+1])
```

#### Iterating:

```python
def result(p: float, maxSteps: int, iterations: int) -> None:
    """
    the function to iterate and repeat getting results in a for loop and plotting mean of the results

    Xs is the list of xs for different iterations
    Vars is the list of vars for different steps
    menaXs is the list of meanXs for different steps

    :param p: the possibility to go to the right(a number between 0 and 1)
    :type p: float
    :param maxSteps: the total steps the walker goes
    :type maxSteps: int
    :param iterations: the number of iterations
    :type iterations: int
    """
    meanXs = []
    varXs = []
    for steps in range(maxSteps):
        xs =[]
        for i in range(iterations):
            rw = RandomWalk(p,steps)
            x = rw.getX(steps)
            xs.append(x)

        meanXs.append(np.mean(xs))
        varXs.append(np.var(xs))


    plt.figure(figsize=(10,10))
    plt.title("mean position of random walk with p ={} vs steps".format(p))
    plt.xlabel("steps")
    plt.ylabel("mean position")
    plt.scatter(np.arange(maxSteps),meanXs)
    plt.show()

    print("r2 for steps and meanXs: ",np.corrcoef(np.arange(maxSteps),meanXs)[0][1])
    print("slope: ",np.polyfit(np.arange(maxSteps),meanXs,1)[0])


    plt.figure(figsize=(10,10))
    plt.title("mean variance of position of random walk with p ={} vs steps".format(p))
    plt.xlabel("steps")
    plt.ylabel("mean var of position")
    plt.scatter(np.arange(maxSteps),varXs)
    plt.show()

    print("r2 for steps and meanXs: ",np.corrcoef(np.arange(maxSteps),varXs)[0][1])
    print("slope: ",np.polyfit(np.arange(maxSteps),varXs,1)[0])
```
